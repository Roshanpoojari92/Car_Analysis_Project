```python
df['Manufacturer'].value_counts()
```

```python
# Treating the missing values

for i in df.columns:
  if df[i].dtypes=='object':
    m=df[i].mode()[0] # modal value of every categorical column
    df[i]=df[i].fillna(m) # fill up null values with mode values
  else:
    me=df[i].mean()
    df[i]=df[i].fillna(me)
```
